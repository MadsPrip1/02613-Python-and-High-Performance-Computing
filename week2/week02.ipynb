```python
def listsum(list):
    return sum(list)

list = [1, 2, 3, 4]
print('Output for listsum', listsum(list)) 
```

```python
def deduplicate(list):
    return [*set(list)]

list = [1, 2, 3, 3, 2, 2, 4]
print('Output for deduplicate', deduplicate(list))
```

```python
def sorttuples(list):
    return sorted(list, key=lambda tupple: tupple[1]) # Use lambda to describe that we want to sort for the last element in the tuple

list = [(2,5), (1,2), (4,4), (2,3), (2,1)]
print('Output for sorttuples', sorttuples(list))
```

```python
def squarecubes(list):
    return ([x**2 for x in list], [x**3 for x in list]) 

list = [1, 2, 3, 4]
print('Output for squarecubes', squarecubes(list))
```

```python
list_of_strings = ['2', '5', '3', '6']
list_of_even_numbers = [x for x in list_of_strings if float(x) % 2 == 0]
print(list_of_even_numbers)
```

```python
class Student:
    def __init__(self, name, courses):
        self.name = name
        self.courses = courses

    def attends(self, courses):
        if courses in self.courses:
            return True
        else:
            return False 
        
s = Student('X', ['01005', '02613'])
print(s.attends('02613'))
print(s.attends('02618'))
```

```python
def coursestudents(list_of_students, course_string):
    students_attends_course = []
    for student in list_of_students:
        if student.attends(course_string):
            students_attends_course.append(student.name)
    return students_attends_course

students = [Student('A', ['01005']), Student('B', ['02613']), Student('C', ['01005', '02613'])]

print(coursestudents(students, '02613'))
```

```python
def magnitude(array):
    return np.linalg.norm(array)

array = np.array([1, 1, 3, 3, 4])

print('Output for magnitude is', magnitude(array))
```
