**a)  What is the parallel fraction of the task consisting of the two subtasks?** 

The parallel fractions is given as: 
$$ f = \frac{T_{parallel}}{T_{total}} = \frac{100}{120} = 0.833$$


```python
f = 100/120
f
```

**b) What is the theoretical speedup for p = 10 processors?**

The theoretical speedup is given as:
$$ S = \frac{1}{(1-f) + \frac{f}{p}} = \frac{1}{(1-0.833) + \frac{0.833}{10}} = 4$$

```python
p = 10
s = 1/((1-f) + f/p)
s
```

**c) What is the theoretical maximum speedup?**

The theoretical maximum speedup is given as:
$$ S = \frac{1}{(1-f)} = \frac{1}{(1-0.833)} = 6$$

```python
s_theory = 1/(1 - f)
s_theory
```
