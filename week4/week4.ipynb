## Boardcasting

### Autolab 1



```python
def standardize_rows(data, mean, std):
    return (data - mean) / std
    
data = np.array([[1, 2, 3], [4, 5, 6]])
mean = np.array([0.5, 1, 3])
std = np.array([1, 2, 3])

print(standardize_rows(data, mean, std))
```

### Autolab 2



```python
def outer(x,y):
    return x[:,None] @ y[None,:]

x = np.array([1,2])
y = np.array([3,4,5])

print('The shape of x is:', x.shape)   
print('The shape of x[None,:] is:', x[None,:].shape)
print('The shape of y is:', y.shape)     
print('The shape of y[None,:] is:', y[None,:].shape)

print(outer(x,y))
```

### Autolab 3



```python
def distmat_1d(x, y):
    return np.abs(x[:, None] - y[None,:])

x = np.array([1, 2])
y = np.array([3, 0.5, 1])

print('The shape of x is:', x.shape)   
print('The shape of x[None,:] is:', x[None,:].shape)
print('The shape of y is:', y.shape)     
print('The shape of y[None,:] is:', y[None,:].shape)

print(distmat_1d(x,y))
    
```
